```python
import pandas as pd

# Load dataset
df = pd.read_csv("bangladesh food price index.csv")
df
```

```python

# Remove metadata row (first row contains column descriptions)
df = df[df["date"] != "#date"].copy()
df
```

```python

# Convert date column to datetime
df["date"] = pd.to_datetime(df["date"], errors="coerce")
```

```python

# Filter only 2024 data
df_2024 = df[df["date"].dt.year == 2024].copy()

# Select and rename required columns
expected_format = df_2024[["commodity", "unit", "price", "currency"]].copy()
expected_format.columns = ["item", "unit", "price", "currency"]

# Optional: reset index
expected_format.reset_index(drop=True, inplace=True)

# Save to new CSV (optional)
expected_format.to_csv("bangladesh_food_prices_2024_formatted.csv", index=False)

print(expected_format.head())
```
